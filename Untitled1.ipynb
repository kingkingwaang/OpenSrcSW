```python
import matplotlib.pyplot as plt

xdata = [1,2,3,4]
ydata = [10,5,20,50]

plt.title("plot_graph")
plt.plot(xdata,ydata)
plt.show()

plt.scatter(xdata,ydata)
plt.show()

ydata = [10,25,37,50]
plt.bar(xdata,ydata)
plt.show()

plt.barh(xdata,ydata)
plt.show()

plt.hist([1,1,2,3,3,3,4,4,4,5,5,5,5,5,5])
plt.show()

data = [10,30,60]
plt.pie(data)
plt.show()
```
